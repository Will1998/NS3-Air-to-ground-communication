```python
import numpy as np
def gps2meters(lat1, long1, lat2, long2):
    latMid = (lat1 + lat2) / 2.0
    m_per_deg_lat = 111132.92 - 559.822 * np.cos(2.0 * latMid) + 1.175 * np.cos(4.0 * latMid) - 0.0023 * np.cos(
        6 * latMid)
    #m_per_deg_lon = (3.14159265359 / 180) * 6367449 * np.cos(latMid)
    m_per_deg_lon = 111412.84 * np.cos(latMid) - 93.5 * np.cos(3 * latMid) + 0.118 * np.cos(5 * latMid)

    deltaLat = (lat1 - lat2)
    deltaLon = (long1 - long2)
    deltax = deltaLat * m_per_deg_lat
    deltay = deltaLon * m_per_deg_lon  # meter
    return deltay,deltax
```

## The trajectory of drone in space-time 3D

### wifi1

```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
import matplotlib.animation as animation
import pandas as pd
from sys import exit
def update_lines(num, data, line):
    # NOTE: there is no .set_data() for 3 dim data...
    line.set_data(data[0:2, :num])    
    line.set_3d_properties(data[2, :num])    
    return line

# Attaching 3D axis to the figure
fig = plt.figure(figsize=(10,6))
ax = p3.Axes3D(fig)

# Reading the data from a CSV file using pandas

repo = pd.read_csv('DJIFlightRecord_2020-10-18_[10-15-42]-TxtLogToCsv.csv', encoding = "ISO-8859-1")
repo = repo[['OSD.longitude','OSD.latitude','OSD.height [m]']]
lon1=repo.iloc[0,0]
lat1=repo.iloc[0,1]
repo["longitude [m]"],repo["latitude [m]"]=gps2meters(lat1,lon1,repo['OSD.latitude'],repo['OSD.longitude'])
data = np.array((repo["longitude [m]"].values, repo['latitude [m]'].values, repo['OSD.height [m]'].values))
data =data[:,::5]
#exit()

# Creating fifty line objects.
line = ax.plot(data[0, 0:1], data[1, 0:1], data[2, 0:1])[0]

x1=repo['longitude [m]'].min()
x2=repo['longitude [m]'].max()
y1=repo['latitude [m]'].min()
y2=repo['latitude [m]'].max()
z1=repo['OSD.height [m]'].min()
z2=repo['OSD.height [m]'].max()

# Setting the axes properties
ax.set_xlim3d([x1,x2])
ax.set_xlabel('X [m]')

ax.set_ylim3d([y1,y2])
ax.set_ylabel('Y [m]')

ax.set_zlim3d([z1,z2])
ax.set_zlabel('height[m]')

ax.set_title('Drone Trajectory-WiFi1')

# Creating the Animation object
line_ani = animation.FuncAnimation(fig, update_lines, data.shape[1], fargs=(data, line), interval=1, blit=False)
#line_ani.save('wifi1.gif', writer='imagemagick', fps=60)


#plt.show()
```
